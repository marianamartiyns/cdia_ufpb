## sem a def de podar, encontrar+proximo e imprimir e sem aquele negocio de 'global'

```python
from mip import Model, MAXIMIZE, xsum
from numpy import asarray, abs
import re

def ler_txt(caminho_arquivo: str):
    l = []
    with open(caminho_arquivo) as arquivo:
        for linha in arquivo:
            for i in re.findall(r'\d+', linha):
                l.append(i)

    variaveis = int(l[0])
    restricoes = int(l[1])
    coef_obj = []
    for i in range(2, 2 + variaveis):
        coef_obj.append(int(l[i]))

    count = 0
    aux = []
    coef_restricoes = []
    for i in range(2 + variaveis, len(l)):
        if count == variaveis:
            aux.append(int(l[i]))
            coef_restricoes.append(aux)
            aux = []
            count = 0
        else:
            aux.append(int(l[i]))
            count += 1
    return variaveis, restricoes, coef_obj, coef_restricoes

def resolver(modelo: Model):
    modelo.optimize()
    parametros = {}
    parametros["objetivo"] = modelo.objective_value
    parametros["variaveis"] = modelo.vars
    return parametros


def ramificar(modelo: Model, valores_solucao):
    var_ramificacao = valores_solucao[encontrar_mais_proximo([i.x for i in valores_solucao], 0.5)]

    modelo_0 = modelo.copy()
    modelo_0 += var_ramificacao == 0

    modelo_1 = modelo.copy()
    modelo_1 += var_ramificacao == 1

    return modelo_0, modelo_1


# podar e encontrar+proximo ta dentro de branch agora
def branch_and_bound(modelo: Model, primal=0, modelo_otimo=None):
    
    nos = [modelo]

    while nos:
        modelo_solver = resolver(nos[0])
        count_int = 0

        if modelo_solver["objetivo"] is None:
            nos.pop(0)
            continue

        for i in modelo_solver["variaveis"]:
            if i.x.is_integer():
                count_int += 1

        if count_int == len(modelo_solver["variaveis"]):
            if modelo_solver["objetivo"] >= primal:
                modelo_otimo = nos[0]
                primal = modelo_solver["objetivo"]
            nos.pop(0)
            continue

        if modelo_solver["objetivo"] <= primal:
            nos.pop(0)
            continue

        # Lógica de poda
        aux_solver = resolver(nos[0])
        count_int = 0

        if aux_solver["objetivo"] is None:
            nos.pop(0)
            continue

        for i in aux_solver["variaveis"]:
            if i.x.is_integer():
                count_int += 1

        if count_int == len(aux_solver["variaveis"]):
            if aux_solver["objetivo"] >= primal:
                modelo_otimo = nos[0]
                primal = aux_solver["objetivo"]
            nos.pop(0)
            continue

        # Lógica de ramificação
        var_ramificacao_idx = encontrar_mais_proximo([i.x for i in modelo_solver["variaveis"]], 0.5)
        var_ramificacao = modelo_solver["variaveis"][var_ramificacao_idx]

        modelo_0 = nos[0].copy()
        modelo_0 += var_ramificacao == 0

        modelo_1 = nos[0].copy()
        modelo_1 += var_ramificacao == 1

        nos.append(modelo_0)
        nos.append(modelo_1)
        nos.pop(0)

    return primal, modelo_otimo

def executar(caminho_arquivo):
    
    variaveis, restricoes, coef_obj, coef_restricoes = ler_txt(caminho_arquivo)

    modelo = Model(sense=MAXIMIZE)

    x = [modelo.add_var(var_type="CONTINUOUS", lb=0, ub=1, name="x_" + str(i)) for i in range(variaveis)]

    modelo.objective = xsum(coef_obj[i] * x[i] for i in range(variaveis))

    for i in range(restricoes):
        modelo += xsum(coef_restricoes[i][j] * x[j] for j in range(variaveis)) <= coef_restricoes[i][-1]

    primal, modelo_otimo = branch_and_bound(modelo)

    modelo_solver = resolver(modelo_otimo)
    if modelo_solver["objetivo"] is not None:
        print("Variáveis:")
        for i in modelo_solver["variaveis"]:
            print(i.name, ' = ', i.x)
        print("Função objetivo:")
        print('Z = ', modelo_solver["objetivo"])
    else:
        print("O problema é inviável.")

# Exemplo de uso:
executar("teste2.txt")
```
