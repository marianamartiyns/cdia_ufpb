## TESTES PARA HOMOGENEIDADE DE VARIÂNCIAS : O Teste χ2 de Bartlett

<fieldset>
    <p>O teste original proposto para verificar a homogeneidade de variâncias entre grupos é o teste χ2 de Bartlett (1937), ou seja, é um método estatístico utilizado para verificar se a <strong>variância</strong> entre vários grupos é igual.</p>
    <p>As hipóteses nulas e alternativas do teste de Bartlett são as seguintes:</p>
<div style="border:2px solid black; padding: 10px; background-color:#CEE0DC">
    <ul>
        <li><strong>H0</strong>: A variância é a mesma para todos os grupos.</li>
        <li><strong>HA</strong>: Pelo menos um grupo tem uma variância que é diferente dos outros.</li>
    </ul>  </div>
    <p>No código:
        <ul>
        <li>Geramos uma lista de amostras aleatórias para cada grupo usando a função np.random.normal para gerar amostras de uma distribuição normal;</li>
        <li>Com np.random.uniform geramos o desvio padrão para cada grupo. </li>
        <li>Em seguida, aplicamos o teste de Bartlett às amostras manualmente e usando a função stats.bartlett da biblioteca scipy.stats. Esta função retorna a estatística de teste e o valor-p.</li>
    </ul>
    
    A estatística de teste é calculada da seguinte forma:
$$B = \frac{(nk)lns^2 - \Sigma (n_j -1)logs_j^2}{c}$$
Onde:
- n: o número total de observações em todos os grupos
- k: O número total de grupos
- ln: Isso significa “log natural”
- s^2: A variação agrupada
- n_j: O número de observações no grupo j
- s_j^2: A variância do grupo j

E c é calculado da seguinte forma:
$$c = 1 + \frac{1}{3 (k-1)}* (\Sigma \frac{1}{n_j -1} - \frac{1}{N - k})$$

  Se o valor-p for menor que 0.05, rejeitamos a hipótese nula. Isso significa que pelo menos um grupo tem uma variância que é diferente dos outros.</p>
</fieldset>

### Exemplo

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd


'''Suponha que somos um pesquisadores e coletamos dados de três diferentes espécies de plantas. 
Queremos saber se a altura das plantas varia significativamente entre as espécies.

-> H0 (Hipótese Nula): A variância da altura das plantas é a mesma para todas as três espécies. 
    Isso significa que não há diferença significativa na variabilidade da altura entre as espécies.

-> HA (Hipótese Alternativa): Pelo menos uma espécie tem uma variância de altura que é diferente das outras. 
    Isso significa que há uma diferença significativa na variabilidade da altura entre pelo menos duas espécies.'''

num_species = k = 3
num_plants = 5000

# Gerando alturas para três espécies diferentes
np.random.seed(0)  # Para reprodutibilidade
species_1 = np.random.normal(loc=150, scale=20, size=num_plants)
species_2 = np.random.normal(loc=160, scale=22, size=num_plants)
species_3 = np.random.normal(loc=155, scale=18, size=num_plants)
```

### Verificação se nossos dados estão normalmente distribuídos

```python
# Função para plotar histograma e Q-Q plot e Verificar se os dados estão normalmente distribuídos
def plot_data(data, title, row):
    # Histograma
    plt.subplot(3, 2, 2*row-1)
    plt.hist(data, bins='auto')
    plt.title(f"Histograma - {title}")

    # Q-Q plot
    plt.subplot(3, 2, 2*row)
    stats.probplot(data, plot=plt)
    plt.title(f"Gráfico Q-Q - {title}")

    # Teste de Shapiro-Wilk para normalidade
    stat, p = stats.shapiro(data)
    print(f"{title} - Estatística de teste: {stat}, Valor-p: {p}")
    if p > 0.05:
        print(f"{title} parece ser normalmente distribuído (p > 0.05).\n")
    else:
        print(f"{title} não parece ser normalmente distribuído (p <= 0.05).\n")

# Criar uma figura grande para acomodar todos os subplots
plt.figure(figsize=(11,16))

# Plotar os dados para cada espécie
plot_data(species_1, "Espécie 1", 1)
plot_data(species_2, "Espécie 2", 2)
plot_data(species_3, "Espécie 3", 3)

# Mostrar todos os plots juntos
plt.tight_layout()
plt.show()
```

---
## Teste χ2 de Bartlett "manualmente"

```python
# Número total de observações
N = len(species_1) + len(species_2) + len(species_3) #1500

# Variâncias dos grupos
s1 = np.var(species_1, ddof=1)
s2 = np.var(species_2, ddof=1)
s3 = np.var(species_3, ddof=1)

# Variância agrupada
s_pooled = ((len(species_1) - 1) * s1 + (len(species_2) - 1) * s2 + (len(species_3) - 1) * s3) / (N - k)

# Estatística do teste
q = (N - k) * np.log(s_pooled) - (len(species_1) - 1) * np.log(s1) - (len(species_2) - 1) * np.log(s2) - (len(species_3) - 1) * np.log(s3)
c = 1 + 1/(3*(k-1)) * ((1/(len(species_1) - 1) + 1/(len(species_2) - 1) + 1/(len(species_3) - 1)) - 1/(N - k))
B = q / c

# Valor-p
p = stats.chi2.sf(B, k - 1)

print(f"Teste de Bartlett - Estatística: {B}, Valor-p: {p}")

if p > 0.05:
    print("\nNão há evidência suficiente para rejeitar a hipótese nula de igualdade de variâncias entre as espécies.")
else:
    print("\nHá evidências para rejeitar a hipótese nula de igualdade de variâncias entre as espécies.")
```

---
## c/  stats.bartlett

```python
# H0: A variância da altura é a mesma para todas as espécies
# HA: Pelo menos uma espécie tem uma variância de altura que é diferente das outras

# Teste de Bartlett para igualdade de variâncias
stat, p = stats.bartlett(species_1, species_2, species_3)
print(f"Teste de Bartlett - Estatística: {stat}, Valor-p: {p}")

if p > 0.05:
    print("\nNão há evidência suficiente para rejeitar a hipótese nula de igualdade de variâncias entre as espécies.")
else:
    print("\nHá evidências para rejeitar a hipótese nula de igualdade de variâncias entre as espécies.")
```

## **Relatório do Teste de Bartlett**

**Contexto da Pesquisa**: Pesquisadores coletaram dados de altura de três espécies de plantas para determinar se há variação significativa entre elas.

**Hipóteses**:
- **H0**: Não há diferença significativa na variabilidade da altura entre as espécies.
- **HA**: Existe diferença significativa na variabilidade da altura entre pelo menos duas espécies.

**Resultados**:
- **Estatística de Teste**: 224.93775262979258
- **Valor-p**: 1.4301698905734775e-49

**Conclusão**: Com base na estatística de teste e no valor-p extremamente baixo, há evidências estatísticas para rejeitar a hipótese nula. Isso indica que a variância da altura das plantas é significativamente diferente entre pelo menos duas das três espécies estudadas. Portanto, a variabilidade da altura das plantas não é uniforme entre as espécies.

```python
data = {
    'Espécie 1': species_1,
    'Espécie 2': species_2,
    'Espécie 3': species_3
}

df = pd.DataFrame(data)

# Calcular desvio padrão e variância para cada espécie (coluna)
desvio_padrao_por_especie = df.std()
variancia_por_especie = df.var()
tamanho_amostra_por_especie = df.count()

# resumos
df.loc['Desvio-padrão'] = desvio_padrao_por_especie
df.loc['Variância'] = variancia_por_especie
df.loc['Tamanho da amostra'] = tamanho_amostra_por_especie

limited_df = pd.concat([df.iloc[:10], df.iloc[-3:]])

fig, ax = plt.subplots(figsize=(10, 3))
table_data = limited_df.round(6).values
col_labels = limited_df.columns.tolist()
row_labels = limited_df.index.tolist()
table = ax.table(cellText=table_data, colLabels=col_labels, rowLabels=row_labels, loc='center')
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1, 1.5)
ax.axis('off')
plt.show()
```
