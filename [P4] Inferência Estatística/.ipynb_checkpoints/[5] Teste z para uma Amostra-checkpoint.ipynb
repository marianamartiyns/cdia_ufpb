# O Teste z para uma Amostra

O teste Z para uma amostra √© utilizado para determinar se a **m√©dia de uma amostra (denotada como (xÃÖ) √© significativamente diferente da m√©dia populacional conhecida (denotada como ùúá)**, sob a *suposi√ß√£o* de que a distribui√ß√£o da popula√ß√£o √© normal. Esse teste √© apropriado quando o tamanho da amostra √© grande (tipicamente, n ‚â• 30) e a vari√¢ncia da popula√ß√£o (œÉ¬≤ 
) √© conhecida. 

<p>As hip√≥teses nulas e alternativas do teste z s√£o as seguintes:</p>
<div style="border:2px solid black; padding: 10px; background-color:#CEE0DC">
    <ul>
        <li><strong>H0</strong>:  A m√©dia da popula√ß√£o √© igual a um valor espec√≠fico. Matematicamente, \(H_0: \mu = \mu_0\), onde \(\mu_0\) √© o valor especificado para a m√©dia populacional.</li><br>
        <li><strong>HA</strong>: Depende do contexto do teste: <br> 
  - Bicaudal: \(H_a: \mu \neq \mu_0\) <br>
  - Unicaudal √† direita: \(H_a: \mu > \mu_0\)<br>
  - Unicaudal √† esquerda: \(H_a: \mu < \mu_0\)</li>
    </ul>  </div>

N√≠vel de signific√¢ncia (Œ±): √â a probabilidade de rejeitar a hip√≥tese nula quando ela √© verdadeira. Comumente, (Œ±) √© escolhido como 0.05.

- Para um teste de duas caudas com (Œ± = 0.05), os valores cr√≠ticos seriam aproximadamente \(\pm1.96\) (dependendo da tabela de distribui√ß√£o normal padr√£o).
- Para um teste unicaudal, o valor cr√≠tico seria \(1.645\) para (Œ± = 0.05) na cauda direita, ou \(-1.645\) na cauda esquerda.

```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro
import scipy.stats as stats

# Fun√ß√£o para realizar o teste Z para uma amostra
def teste_z_amostra(media_amostra, media_populacional, desvio_populacional, tamanho_amostra, alpha, tipo_teste='bicaudal'):
    
    # Calcula a estat√≠stica de teste Z
    z = (media_amostra - media_populacional) / (desvio_populacional / np.sqrt(tamanho_amostra))
    
    # Determina o valor cr√≠tico com base no tipo de teste
    if tipo_teste == 'bicaudal':
        valor_critico = norm.ppf(1 - alpha / 2)
        regiao_critica = (-valor_critico, valor_critico)
    elif tipo_teste == 'unicaudal_direita':
        valor_critico = norm.ppf(1 - alpha)
        regiao_critica = (valor_critico, np.inf)
    elif tipo_teste == 'unicaudal_esquerda':
        valor_critico = norm.ppf(alpha)
        regiao_critica = (-np.inf, valor_critico)
    else:
        raise ValueError("Tipo de teste inv√°lido. Escolha entre 'bicaudal', 'unicaudal_direita' ou 'unicaudal_esquerda'.")
    
    # Decide se rejeita ou n√£o a hip√≥tese nula
    if z < regiao_critica[0] or z > regiao_critica[1]:
        rejeitar_hipotese_nula = True
    else:
        rejeitar_hipotese_nula = False
    
    return z, regiao_critica, rejeitar_hipotese_nula

# Gerando uma base de dados aleat√≥ria para exemplo
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
tamanho_amostra = 5000
media_populacional = 10
desvio_populacional = 2
amostra = np.random.normal(loc=media_populacional, scale=desvio_populacional, size=tamanho_amostra)

# Calculando a m√©dia da amostra
media_amostra = np.mean(amostra)

# Realizando o teste de Shapiro-Wilk para verificar normalidade dos dados
stat_test_shapiro, p_value_shapiro = shapiro(amostra)

alpha = 0.05 # Definindo o n√≠vel de signific√¢ncia (alpha)

# Verificando a normalidade dos dados
if p_value_shapiro > alpha:
    print("Os dados parecem seguir uma distribui√ß√£o normal (n√£o rejeitamos a hip√≥tese nula do teste de Shapiro-Wilk)")
else:
    print("Os dados n√£o parecem seguir uma distribui√ß√£o normal (rejeitamos a hip√≥tese nula do teste de Shapiro-Wilk)")
      
def plot_data(data, title):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins='auto')
    plt.title(f"Histograma das Alturas - {title}")

    plt.subplot(1, 2, 2)
    stats.probplot(data, plot=plt)
    plt.title(f"Gr√°fico Q-Q das Alturas - {title}")

    plt.show()
    
plot_data(amostra, "Amostra")
```

```python
# Realizando o teste Z para uma amostra
z, regiao_critica, rejeitar_hipotese_nula = teste_z_amostra(media_amostra, media_populacional, desvio_populacional, tamanho_amostra, alpha, tipo_teste='bicaudal')

# Criando um DataFrame para mostrar os dados
dados = pd.DataFrame({
    'Amostra': amostra,
})

# Imprimindo os resultados
print("\nDados da Amostra:")
print(dados)

print("\nResultados do Teste Z:")
print(f"\nEstat√≠stica de teste Z: {z}")
print(f"\nRegi√£o cr√≠tica: {regiao_critica}")
print(f"\nRejeitar hip√≥tese nula? {rejeitar_hipotese_nula}")
```

- Se a vari√¢ncia populacional n√£o for conhecida e a amostra for grande, o desvio padr√£o da amostra pode ser usado como uma aproxima√ß√£o de \(\sigma\).
- Para amostras menores, o teste t de Student √© mais apropriado devido √† variabilidade maior nas estimativas de \(\sigma\) a partir de amostras pequenas.
