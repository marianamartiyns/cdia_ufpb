**QUESTÃO 1**: Escreva um programa tal que dados dois números naturais Kmin e Kmax, com Kmax − Kmin ≥ 3 e que exiba numa mesma figura o gráfico de uma funÇão f(x) genérica num intervalo contendo um número real a e os gráficos das (k + 1) somas parciais (polinômios de Taylor) P0(x, a), P1(x, a), P2(x, a), · · · , Pk(x, a) da série de Taylor de f(x), considerando x num intervalo [xmin, xmax] conveniente contendo o centro a da série. 

Fornece como dados de entrada: 

(a) a expressão da função f(x); <BR>
(b) o centro a da serie de potencias; <BR>
(c) os números naturais Kmin e Kmax, com Kmax − Kmin ≥ 3, que definem as somas parciais a serem plotadas. <BR>
(d) os números reais xmin e xmax, com xmin < a < xmax, que define o intervalo onde os gráficos das somas parciais será plotado. Caso o raio R de convergência da série seja finito (não tão grande) procure considerar xmin < a − R e xmax > a + R, e considere também um n´umero real δ > 0, porém suficientemente pequeno, para que o grafico da função f seja exibido apenas para x no intervalo fechado [a − R + δ, a + R − δ] ⊂ (a − R, a + R) diferentemente dos gráficos das somas parciais que devem ser exibidos no intervalo [xmin, xmax] contendo o intervalo de convergência da série (a − R, a + R). <BR><BR>

Ilustre figuras geradas pelo programa, identificando qual a função f(x) e qual o centro a da série considerados.


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)  # Exemplo de função

# função f(x) em torno do ponto a
def taylor_series(f, a, x, k):
    series = 0
    for n in range(k + 1):
        series += (f(a) / np.math.factorial(n)) * (x - a) ** n
    return series


a = np.pi / 2  # centro da série
k_min = 0      # Valor mínimo de k
k_max = 5      # Valor máximo de k
xmin = a - np.pi  # Valor mínimo de x
xmax = a + np.pi  # Valor máximo de x


x_values = np.linspace(xmin, xmax, 1000)
plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label='f(x) = sin(x)', color='black')

#somas parciais da série de Taylor
for k in range(k_min, k_max + 1):
    plt.plot(x_values, taylor_series(f, a, x_values, k), label=f'P{k}(x, π/2)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Somas Parciais da Série de Taylor para f(x) = sin(x)')
plt.axvline(x=a, color='r', linestyle='--', label='Centro da Série')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)  # Exemplo de função

# Função para calcular a série de Taylor de f(x) em torno do ponto a
def taylor_series(f, a, x, k):
    series = 0
    for n in range(k + 1):
        series += (f(a) / np.math.factorial(n)) * (x - a) ** n
    return series

'''
- A função f(x) é sin(x) sin(x).
- O centro da da série de potências é π/2.
- Os números naturais Kmin e Kmax são 0 e 5.
- Os números reais xmin e xmax são a−2 e a+2, respectivamente. '''

a = np.pi / 2  
Kmin = 0       
Kmax = 5    
xmin = a - 2  
xmax = a + 2   

# Criação dos valores de x para plotagem
x_values = np.linspace(xmin, xmax, 1000)
plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label='f(x) = sin(x)', color='black')

# Plot das somas parciais da série de Taylor
for k in range(Kmin, Kmax + 1):
    plt.plot(x_values, taylor_series(f, a, x_values, k), label=f'P{k}(x, π/2)')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Somas Parciais da Série de Taylor para f(x) = sin(x)')
plt.axvline(x=a, color='r', linestyle='--', label='Centro da Série')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# f(x) = 1/(1+x^2)
def f(x):
    return 1 / (1 + x ** 2)

# polinômio
coeficientes = [1/np.math.factorial(0), 0/np.math.factorial(1), -2/np.math.factorial(2), 0/np.math.factorial(3), 24/np.math.factorial(4)]

# intervalo de x
x = np.linspace(-100, 100, 400)


y_polynomial = np.polyval(coefficients[::-1], x)
y_function = f(x)

# Plotando os gráficos
plt.figure(figsize=(10, 6))
plt.plot(x, y_polynomial, label='Polinômio', color='blue')
plt.plot(x, y_function, label='$f(x)=\\frac{1}{1+x^2}$', color='red')

plt.title('Gráficos do Polinômio e da Função $f(x)=\\frac{1}{1+x^2}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
