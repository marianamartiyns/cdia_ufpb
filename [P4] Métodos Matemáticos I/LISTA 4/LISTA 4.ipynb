### Questão 2

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_sequencia(nmin, nmax, epsilon):
    S = [0]
    for n in range(1, nmax+1):
        Sn = S[-1] + 1/n
        S.append(Sn)
        if n >= nmin and abs(S[-1] - S[-2]) < epsilon:
            break
    return S[nmin-1:n+1]

def exibir_tabela_e_grafico(nmin, nmax, S):
    print(f"{'n':>5} {'Sn':>10}")
    for n, Sn in enumerate(S, start=nmin):
        print(f"{n:>5} {Sn:>10.6f}")
    
    plt.figure(figsize=(10, 6))
    plt.plot(range(nmin, nmin+len(S)), S, marker='o')
    plt.xlabel('n')
    plt.ylabel('Sn')
    plt.title('Sequência Sn')
    plt.grid(True)
    plt.show()

nmin = 100  # Valor mínimo de n
nmax = 110  # Valor máximo de n
epsilon = 0.001  # Tolerância

S = calcular_sequencia(nmin, nmax, epsilon)
exibir_tabela_e_grafico(nmin, nmax, S)
```
