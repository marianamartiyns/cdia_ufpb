```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    
    #construtor
    def __init__(self):
        self.head = None
        self._size = 0
        
    #inserir elemento no final da lista
    def append (self, elem):
    
        if self.head: #inserção quando a lista já possui elementos
            
            # pointer é uma variavel auxiliar
            pointer = self.head
            while(pointer.next):
                pointer = pointer.next
            pointer.next = Node(elem)
            
        else:
            self.head = Node(elem) # primeira inserção
        
        self._size +=1
        
    def __len__(self): #retorna o tamanho da lista
        return self._size
    
    def __getitem__(self, index): # a = lista.get[6]
        
        pointer = self._getnode(index)
        
        if pointer:
            return pointer.data
        raise InderError("list index out of range")
        
    
    def __setitem__(self, index, elem): # lista.set[5] = 9
        
        pointer = self._getnode(index)
        
        if pointer:
            pointer.data = elem
        else:
            raise InderError("list index out of range")
    
    def _getnode(self, index):
        
        pointer = self.head
        for i in range(index):
            if pointer: #verificando se o pointer não é vazio
                pointer = pointer.next
            else:
                raise InderError("list index out of range") 
                
        if pointer:
            return pointer.data
        raise InderError("list index out of range")
        
        return pointer
        
    # busca linear
    def index(self, elem): #retorna o indice de um elemento da lista
    
        pointer = self.head
        i = 0
        while(pointer): #enauqnto o pointer tiver um nó que é difente do vazio
            if pointer.data == elem:
                return i
            pointer = pointer.next
            i += 1
            
        raise ValueError(f"{elem} is not in list")
     
    # Inserir elemento na lista
    def insert(self, index, elem):
        
        node = Node(elem)
        if index == 0: # quer inserir o elemento na head da lista
            node.next = self.head
            self.head = node
            
        else: 
            pointer = self._getnode(index-1)
            node.next = pointer.next
            pointer.next = node
            
        self._size += 1
      
    # Remover elemento da lista
    def remove(self, elem):
        if self.head == None:
            ValueError(f"{elem} is not in list")
            
        elif self.head.data == elem:
            self.head = self.head.next
            self._size -= 1
            return True
            
        else:
            ancestor = self.head
            pointer = self.head.next
            while(pointer):
                if pointer.data == elem: # remove
                    
                    ancestor.next = pointer.next
                    pointer.next = None
                    
                ancestor = pointer
                pointer = pointer.next
            self._size -= 1
            return True
        
        ValueError(f"{elem} is not in list")
        
    def __repr__(self):
        r = ""
        pointer = self.head
        while(pointer):
            r += str(pointer.data) + " -> "
            pointer = pointer.next
            
        return r
        
    def __str__(self):
        return self._repr__()
                    
```

## FILA

```python
class queue:
    
    #construtor
    def __init__(self):
        self.first = None
        self.last = None
        self._size = 0
    
    #inserir (apenas no final)
    def push(self, elem):
        
        node = Node(elem)
        if self.last is None:
            self.last = None
        else:
            self.last.next = node
            self.last = node
            
        if self.first is None:
            self.first = node
            
        self._size += 1
        
            
    #remove o primeiro elemento da fila
    def pop(self):
        
        if self.first > 0:
            
            elem = self.first.data
            self.first = self.first.next
            self._size -= 1
            return elem
        
        raise InderError("the queue is empty")
        
    #retorna o primeiro elemento da fila sem remover
    def peek(self):
        
        if self._size > 0:
            elem = self.first.data
            return elem
        raise InderError("the queue is empty")
        

    def __len__(self): #retorna o tamanho da lista
        return self._size

    
    def __repr__(self):
        
        if self._size > 0: 
            r = ""
            pointer = self.first
            while(pointer):
                r += str(pointer.data) + " "
                pointer = pointer.next

            return r
        return "Empty Queue"
        
    def __str__(self):
        return self._repr__()
```
