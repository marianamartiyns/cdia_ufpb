- A média ponderada é calculada somando o produto da quantidade e da frequência para cada categoria e dividindo pela soma total das frequências.
- A mediana é estimada considerando a frequência acumulada e encontrando a categoria que ultrapassa ou iguala a 50% da frequência total.
- A moda é determinada pela categoria com a maior frequência.

```python
import statistics

# Dados agrupados
dados_agrupados = {
    "D": {"quantidade": 137, "frequencia": 95},
    "VQ": {"quantidade": 7, "frequencia": 5},
}

# Calcular média ponderada: Média Ponderada: Indica a tendência central dos 
# dados considerando a importância relativa (frequência) de cada categoria.
media_ponderada = sum(info["quantidade"] * info["frequencia"] for info in dados_agrupados.values()) / 100
print("Média Ponderada:", media_ponderada)

# Calcular mediana
# Como os dados estão agrupados, não podemos calcular a mediana diretamente, 
# mas podemos fazer uma estimativa usando a frequência acumulada
mediana_percentual = 50  # Por exemplo, 50% para mediana
frequencia_acumulada = 0
mediana = None
for categoria, info in dados_agrupados.items():
    frequencia_acumulada += info["frequencia"]
    if frequencia_acumulada >= mediana_percentual:
        mediana = categoria
        break
print("Mediana Estimada:", mediana)

# Calcular moda
# A moda é a categoria com a maior frequência
moda = max(dados_agrupados, key=lambda x: dados_agrupados[x]["frequencia"])
print("Moda:", moda)
```

```python
# Importar as bibliotecas necessárias
import pandas as pd
import scipy.stats as stats

# Criar um DataFrame para representar os dados da pesquisa
data = {
    'Satisfaction': ['Insatisfeito']*52 + ['Razoavelmente satisfeito']*53 + ['Satisfeito']*39,
    'Frequency': ['Diariamente ou De vez em quando']*137 + ['Raramente ou Nunca']*7
}

df = pd.DataFrame(data)

# Criar uma tabela de contingência
crosstab = pd.crosstab(df['Satisfaction'], df['Frequency'])

# Realizar o teste de qui-quadrado
chi2, p, dof, expected = stats.chi2_contingency(crosstab)

# Imprimir a estatística de teste e o valor-p
print(f"\nTeste de Qui-Quadrado - \nEstatística de teste: {chi2}, Valor-p: {p}")

# Interpretar o resultado
if p > 0.05:
    print("\nNão há evidência suficiente para rejeitar a hipótese nula de independência entre as duas variáveis.")
else:
    print("\nHá evidências para rejeitar a hipótese nula de independência entre as duas variáveis.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados da Questão 7
respostas_q7 = ['Insatisfeito', 'Razoavelmente satisfeito', 'Satisfeito']
quantidades_q7 = [46, 51, 47]

# Dados da Questão 26
respostas_q26 = ['Pouco Impacto', 'Moderado', 'Impactam Muito']
quantidades_q26 = [10, 10, 124]

# Criando o intervalo para as barras
x = np.arange(len(respostas_q7))

# Largura das barras
width = 0.35

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando as barras para a Questão 7
rects1 = ax.bar(x - width/2, quantidades_q7, width, label='Questão 7', color='#7F7FFF')

# Plotando as barras para a Questão 26
rects2 = ax.bar(x + width/2, quantidades_q26, width, label='Questão 26', color='r')

# Adicionando rótulos e título
ax.set_ylabel('Quantidade de Respostas')
ax.set_title('Avaliação das Oportunidades de Experiência Prática e Impacto dos Espaços de Recreação')
ax.set_xticks(x)
ax.set_xticklabels(respostas_q7)
ax.legend()

# Função para adicionar rótulos nas barras
def autolabel(rects):
    """Anexa um rótulo acima de cada barra, exibindo sua altura."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # Deslocamento vertical
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adicionando os rótulos nas barras
autolabel(rects1)
autolabel(rects2)

# Exibindo o gráfico
plt.xticks(rotation=45)
plt.show()
```
